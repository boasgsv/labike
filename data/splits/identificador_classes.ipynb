```python
import pandas as pd
import os
```

```python
directory = '.'
aspects = ['collectionSpecie', 'collectionType', 'bioActivity', 'name', 'collectionSite']
```

```python
if os.path.isdir(directory):
    for i in aspects:
        combined_df = pd.DataFrame()
        for filename in os.listdir(directory):
            if filename.endswith(".csv") and i in filename:
                print(filename, i)
                df = pd.read_csv(filename)
                combined_df = pd.concat([combined_df, df], ignore_index=True)
                combined_df.to_csv(i+".csv")
```

```python
vetor = []
valores = dict()
for i in aspects:
    for filename in os.listdir(directory):
        if filename.startswith(i):
            df = pd.read_csv(filename)
            valores[i] = list(df['neighbor'].values)
```
